```python
import pandas as pd

# Assuming Among_Us is your DataFrame

# Grouping by 'Cartoon' instead of 'Atmosphere'
Among_Us_groupings = Among_Us.groupby('Cartoon')

# Getting data for specific cartoons
Scooby_Doo_Among_Us = Among_Us_groupings.get_group('Scooby Doo')
Tom_and_Jerry_Among_Us = Among_Us_groupings.get_group('Tom and Jerry')

```

```python
type(Scooby_Doo_Among_Us)
```

```python
# Grouping by 'Cartoon'
Among_Us_groupings = Among_Us.groupby('Atmosphere')

# Getting data for specific cartoons
Indoor_Among_Us = Among_Us_groupings.get_group('Indoors')
Outdoor_Among_Us = Among_Us_groupings.get_group('Outdoors')

# Plotting histograms for specific cartoon groups
plt.hist(Indoor_Among_Us['Cartoon'], bins=5, label='Indoors', alpha=0.5)
plt.hist(Outdoor_Among_Us['Cartoon'], bins=5, label='Outdoors', alpha=0.5)

plt.xlabel('Cartoon')
plt.ylabel('Frequency')
plt.title('Distribution of Atmosphere with Cartoons')
plt.legend()
plt.show()
```

```python
#People who liked powerpuff girls liked going outdoors, people who liked tom and jerry liked
#the indoors
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming Among_Us is your DataFrame

# Grouping by 'Movie'
Among_Us_groupings = Among_Us.groupby('Movie')

# Getting data for specific movies (Barbie and Oppenheimer)
Barbie_Among_Us = Among_Us_groupings.get_group('Barbie')
Oppenheimer_Among_Us = Among_Us_groupings.get_group('Oppenheimer')

# Plotting histograms for specific movie groups with 'Atmosphere' using 'bar' layout
plt.hist(Barbie_Among_Us['Atmosphere'], bins=5, label='Barbie', alpha=0.5, align='left', rwidth=0.4)
plt.hist(Oppenheimer_Among_Us['Atmosphere'], bins=5, label='Oppenheimer', alpha=0.5, align='right', rwidth=0.4)

plt.xlabel('Atmosphere')
plt.ylabel('Frequency')
plt.title('Distribution of Atmosphere for Movies (Barbie and Oppenheimer)')
plt.legend()
plt.show()
```

```python
#those who liked barbie liked the indoors and those who like oppenheimer liked the outdoors more 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming Among_Us is your DataFrame

# Grouping by 'Movie'
Among_Us_groupings = Among_Us.groupby('Kind')

# Getting data for specific movies (Barbie and Oppenheimer)
Horror_Among_Us = Among_Us_groupings.get_group('Watching a horror film')
Serial_Among_Us = Among_Us_groupings.get_group('Watching documentaries about serial killers')

# Plotting histograms for specific movie groups with 'Atmosphere' using 'bar' layout
plt.hist(Horror_Among_Us['Atmosphere'], bins=5, label='Watching a horror film', alpha=0.5, align='left', rwidth=0.4)
plt.hist(Serial_Among_Us['Atmosphere'], bins=5, label='Watching documentaries about serial killers', alpha=0.5, align='right', rwidth=0.4)

plt.xlabel('Atmosphere')
plt.ylabel('Kind of Movie')
plt.title('Distribution of Atmosphere for Kind of Movies')
plt.legend()
plt.show()
```

```python
#Those who liked watching horror films prefered to stay indoors and those who liked watching documentaries about serial killers liked going outdoors
```
